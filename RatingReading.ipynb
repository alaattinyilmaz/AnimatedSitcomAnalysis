```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
def plotRatings(fileName):
    rat = pd.read_csv(fileName, sep=',', header=None)
    rat.columns = ["season", "episode", "rating", "numVotes"]
    
    rat['season'] = pd.to_numeric(rat['season'])
    rat['episode'] = pd.to_numeric(rat['episode'])
    rat['rating'] = pd.to_numeric(rat['rating'])
    rat['numVotes'] = pd.to_numeric(rat['numVotes'])
    
    totalVote = rat['rating'] * rat['numVotes']
    
    plot = plt.plot(totalVote)
    plt.ylabel('Total Rating')
    plt.xlabel('Number of Episodes')
    
    return plot
```

```python
def plotRatingsOLRemoved(fileName):
    rat = pd.read_csv(fileName, sep=',', header=None)
    rat.columns = ["season", "episode", "rating", "numVotes"]
    
    rat['season'] = pd.to_numeric(rat['season'])
    rat['episode'] = pd.to_numeric(rat['episode'])
    rat['rating'] = pd.to_numeric(rat['rating'])
    rat['numVotes'] = pd.to_numeric(rat['numVotes'])
    
    totalVote = rat['rating'] * rat['numVotes']
    
    elements = np.array(totalVote)
    mean = np.mean(elements, axis=0)
    sd = np.std(elements, axis=0)
    final_list = [x for x in totalVote if (x > mean - 2 * sd)]
    final_list = [x for x in final_list if (x < mean + 2 * sd)]

    plot = plt.plot(final_list)
    plt.ylabel('Total Rating')
    plt.xlabel('Number of Episodes')

    
    return plot
```

```python
rsp = plotRatings("rsp.txt")
plt.title('South Park ratings over time')
plt.show(rsp)
```

```python
rspx = plotRatingsOLRemoved("rsp.txt")
plt.title('South Park ratings over time (outliers removed)')
plt.show(rspx)
```
